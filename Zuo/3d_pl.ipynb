```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
```

```python
# Load the data
full_nl1 = np.load("full_nl1.npy")
full_nl2 = np.load("full_nl2.npy")
full_nl3 = np.load("full_nl3.npy")
nl1 = np.load("nl1.npy")
nl2 = np.load("nl2.npy")
nl3 = np.load("nl3.npy")
```

```python
# Initialize figure with adjusted size
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel(r'$n_1 + l_1$')
ax.set_ylabel(r'$n_2 + l_2$')
ax.set_zlabel(r'$n_3 + l_3$')
ax.set_xlim(0, 20)
ax.set_ylim(0, 20)
ax.set_zlim(0, 20)
# ax.view_init(elev=7, azim=-103)
ax.view_init(elev=14, azim=142)

depth = np.sqrt(nl1**2 + nl2**2 + nl3**2)

for x, y, z in [(nl1, nl2, nl3), (nl1, nl3, nl2), (nl2, nl1, nl3), 
                (nl2, nl3, nl1), (nl3, nl1, nl2), (nl3, nl2, nl1)]:
    scatter = ax.scatter(x, y, z, c=depth, cmap='Accent', s=4, alpha=0.6)

ax.set_xticks([10, 20])
ax.set_yticks([10, 20])
ax.set_zticks([10, 20])

# apex_index = np.argmax(depth)
# apex = (nl1[apex_index], nl2[apex_index], nl3[apex_index])

# # Define the specific points with combinations of min and max values
# min_nl1, min_nl2, min_nl3 = nl1.min(), nl2.min(), nl3.min()
# max_nl1, max_nl2, max_nl3 = nl1.max(), nl2.max(), nl3.max()

# corner1 = (min_nl1, min_nl2, min_nl3)
# corner2 = (min_nl1, min_nl2, max_nl3)
# corner3 = (max_nl3, min_nl1, min_nl2)
# corner4 = (min_nl1, max_nl3, min_nl2)

# for corner in [corner1, corner2, corner3, corner4]:
#     ax.plot([apex[0], corner[0]], [apex[1], corner[1]], [apex[2], corner[2]], color='black', lw=2)

# ax.scatter(*apex, color='lime', s=6)
# ax.scatter(*corner1, color='black', s=6)
# ax.scatter(*corner2, color='black', s=6)
# ax.scatter(*corner3, color='black', s=6)
# ax.scatter(*corner4, color='black', s=6)

# ax.plot([corner1[0], corner2[0]], [corner1[1], corner2[1]], [corner1[2], corner2[2]], color='black', lw=1)
# ax.plot([corner2[0], corner3[0]], [corner2[1], corner3[1]], [corner2[2], corner3[2]], color='black', lw=1)
# ax.plot([corner3[0], corner4[0]], [corner3[1], corner4[1]], [corner3[2], corner4[2]], color='black', lw=1)
# ax.plot([corner4[0], corner1[0]], [corner4[1], corner1[1]], [corner4[2], corner1[2]], color='black', lw=1)
# ax.plot([corner1[0], corner3[0]], [corner1[1], corner3[1]], [corner1[2], corner3[2]], color='black', lw=1)

cbar = fig.colorbar(scatter, ax=ax, orientation='horizontal', pad=0.05, shrink=0.8)
cbar.set_label("Distance from Origin")
ax.set_title("#active = 1000")

plt.tight_layout()
plt.subplots_adjust(top=0.85, bottom=0.15, left=0.15, right=0.9)

plt.savefig("A1000.pdf", format="pdf", bbox_inches="tight",pad_inches=0.3)

plt.show()
```
